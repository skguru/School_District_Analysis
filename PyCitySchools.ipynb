```python
#add each prefix and suffix to remove to a list
prefix_suffix =["Dr. ", "Mr. ", "Mrs. ", "Miss ", " MD", " DDS", " DVM", " Ph.D", " PhD", "Ms. "]
```

```python
#Identify and replace with empty space

for word in prefix_suffix:
    student_data_df["student_name"] = student_data_df["student_name"].str.replace(word, "")
```

```python
#Combine the data into a single data set, use merge function

student_complete_df = pd.merge(student_data_df, school_data_df, on=["school_name", "school_name"])
student_complete_df
```

```python
#Identify school counts using the combined data frame; use unique function
school_count = len(student_complete_df["school_name"].unique())
school_count
```

```python
#Total budget using complete data frame - will add all rows which is not correct.
total_budget = student_complete_df["budget"].sum()
total_budget
```

```python
#total budget using school data
total_budget = school_data_df["budget"].sum()
total_budget
```

```python
#Average reading scores using student data
average_reading_score1 = student_data_df["reading_score"].mean()
average_reading_score1

```

```python
#Average reading scores using student complete merged data; answer will be identical
average_reading_score = student_complete_df["reading_score"].mean()
average_reading_score
```

```python
#Average Math score using complete merged data
average_math_score = student_complete_df["math_score"].mean()
average_math_score
```

```python
#passing math and reading - create data frames that has only those who have passed
passing_math = student_complete_df[student_complete_df["math_score"]>=70]
passing_reading = student_complete_df[student_complete_df["reading_score"]>=70]
passing_math.head()
```

```python
passing_reading.head()

```

```python
#number of students passing math
passing_math_count = passing_math["student_name"].count()
passing_math_count
```

```python
#number of students passing reading
passing_reading_count= passing_reading["student_name"].count()
passing_reading_count
```

```python
#Percentage of students passing math
passing_math_percentage = passing_math_count / float(student_count) *100
passing_math_percentage
```

```python
#Percentage of students passing reading
passing_reading_percentage = passing_reading_count / float(student_count) *100
passing_reading_percentage
```

```python
#Overall passing percentage
overall_passing_percentage = (passing_reading_percentage+passing_math_percentage)/2
overall_passing_percentage
```

```python
#Adding a list of values with keys to create new data frame, hint: dictionary

district_summary_df = pd.DataFrame(
    [{"Total Schools": school_count, "Total Students": student_count, "Total Budget": total_budget, "Average Math Score": average_math_score, "Average Reading Score": average_reading_score, "%Passing Math": passing_math_percentage,
      "%Passing Reading": passing_reading_percentage, "%Overall Passing": overall_passing_percentage}])

district_summary_df = district_summary_df[["Total Schools", "Total Students", "Total Budget", "Average Math Score", "Average Reading Score", "%Passing Math", "%Passing Reading", "%Overall Passing"]]
district_summary_df
```

```python
# Format the "Total Students" to have the comma for a thousands separator.
district_summary_df["Total Students"] = district_summary_df["Total Students"].map("{:,}".format)

district_summary_df["Total Students"]
```

```python
# Format "Total Budget" to have the comma for a thousands separator, a decimal separator, and a "$".

district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.2f}".format)

district_summary_df["Total Budget"]
```

```python
# Format the columns.
district_summary_df["Average Math Score"] = district_summary_df["Average Math Score"].map("{:.1f}".format)

district_summary_df["Average Reading Score"] = district_summary_df["Average Reading Score"].map("{:.1f}".format)

district_summary_df["%Passing Math"] = district_summary_df["%Passing Math"].map("{:.0f}".format)

district_summary_df["%Passing Reading"] = district_summary_df["%Passing Reading"].map("{:.0f}".format)

district_summary_df["%Overall Passing"] = district_summary_df["%Overall Passing"].map("{:.0f}".format)
```
